# Bar Sommelier Algorithmic Evaluation & Flavor-Vector Plotting

This notebook verifies the recommendation engine's mathematical constraints. It simulates cosine similarity scoring between user taste profile affinities and various drink flavor vectors, including vibe modulations and cold-start blending.

```python
import numpy as np
import matplotlib.pyplot as plt

# Canonical Flavor Dimensions
FLAVORS = ["boozy", "sweet", "sour", "bitter", "smoky", "herbal", "fruity", "creamy", "spicy", "refreshing"]

# Define mock drinks
drinks = {
    "Old Fashioned": np.array([8, 4, 2, 6, 2, 2, 2, 2, 2, 2]), # boozy/bitter
    "Paloma":        np.array([4, 4, 7, 2, 2, 2, 6, 2, 2, 8]), # sour/fruity/refreshing
    "Smoky Mezcal":  np.array([5, 4, 6, 2, 9, 2, 2, 2, 2, 4]), # smoky
    "Virgin Mojito":  np.array([0, 5, 6, 2, 0, 4, 4, 2, 0, 8]), # refreshing/zero-proof
}

def cosine_similarity(v1, v2):
    dot_prod = np.dot(v1, v2)
    norm_1 = np.linalg.norm(v1)
    norm_2 = np.linalg.norm(v2)
    if norm_1 == 0 or norm_2 == 0:
        return 0
    return dot_prod / (norm_1 * norm_2)

print("Cosine Similarity Matrix:")
for name1, vec1 in drinks.items():
    for name2, vec2 in drinks.items():
        if name1 != name2:
            sim = cosine_similarity(vec1, vec2)
            print(f"{name1} <-> {name2}: {sim:.3f}")
```

```python
# Plot the flavor vectors for verification
plt.figure(figsize=(10, 6))
x = np.arange(len(FLAVORS))
width = 0.2

for i, (name, vec) in enumerate(drinks.items()):
    plt.bar(x + (i - 1.5) * width, vec, width, label=name)

plt.title("Cocktail Flavor Vectors (0-10)")
plt.xticks(x, FLAVORS)
plt.ylabel("Intensity")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
